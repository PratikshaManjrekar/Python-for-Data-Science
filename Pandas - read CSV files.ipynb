# CSV

```python
from io import StringIO, BytesIO
```

```python
data=('Col1,Col2,Col3\n'
          'a,b,1\n'
          'c,d,2\n'
          'e,f,3\n')
```

```python
type(data)
```

```python
df2=pd.read_csv(StringIO(data))
df2
```

```python
df2=pd.read_csv(StringIO(data),usecols=['Col1','Col3'])
df2
```

```python
df2['Col3'][2]
```

```python
data1=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11\n')
```

```python
df=pd.read_csv(StringIO(data1),dtype=object)
df
```

```python
type(data1)
```

```python
print(type(df['a'][1]))
print(type(df['c']))
```

```python
df=pd.read_csv(StringIO(data1),dtype=float)
df
```

```python
df['b']
```

```python
type(df['b'][1])
```

```python
# converting cols into different datatypes

df=pd.read_csv(StringIO(data1),dtype={'b':float,'c':int,'a':'Int64'})
df
```

```python
df['a'][1]
```

```python
print(type(df['a'][1]))
print(type(df['c'][2]))
print(type(df['b'][0]))
```

```python
df.dtypes
```

```python
data3=('index,a,b,c\n'
      '4,apple,bat,4.5\n'
      '6,cat,dog,5.76')
```

```python
pd.read_csv(StringIO(data3))
```

```python
pd.read_csv(StringIO(data3),index_col=0)
```

```python
pd.read_csv(StringIO(data3),index_col=1)
```

```python
data4=('a,b,c\n'
          '4,apple,bat\n'
           '6,cat,dog')
```

```python
print(data4)
```

```python
pd.read_csv(StringIO(data4))
```

```python
pd.read_csv(StringIO(data4),index_col=False)
```

```python
# Combining usecols and index cols

pd.read_csv(StringIO(data4),index_col=False, usecols=['a','c'])
```

```python
# Quoting and escaped characters - Useful in NLP

data='a,b\n "Hello \\"Bob\\" nice to meet you",6'
df=pd.read_csv(StringIO(data),escapechar='\\')
df
```

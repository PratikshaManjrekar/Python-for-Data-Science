# Seaborn
## - If we have many features in our dataset then seaborn will help us to analyze it
## - If we have only 1 feature lets say, f1 -> univariate
## - f1,f2 -> bivariate
## - if we have more than 2 features we'll need to create 3d or 4d diagrams, but creating a 4d 
##    diagram is not possible in real world so we'll use pairplot


# Distribution plots
## • distplot - represents univariate distribution of data
## • joinplot - will help us to analyse bivariate (2D)
## • pairplot - 4d 

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
# here tip is the dependent feature and all other are independent features
```

```python
df.dtypes
```

## Correlation with heatmap - helps in finding interrelation between features

```python
df.corr()
```

```python
 sns.heatmap(df.corr())
```

# Jointplot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')        # try using different 'kind'   reg-regression line
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde')
```

# Pair plot
# Used when you have more than 2 features
# Also known as scattered plot
# Feature values needs to be integer or float

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

# Dist plot
# - Helps to create histogram

```python
sns.displot(df)
```

```python
sns.distplot(df['tip'])     # here by default kde is True so the density will be counted in percentage
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
sns.distplot(df['total_bill'])
```

# Reading Excel File

```python
excel=pd.DataFrame(np.arange(1,21).reshape(4,5),index=['Row1','Row2','Row3','Row4'],columns=['Col1','Col2','Col3','Col4','Col5'])
excel.head()
```

```python
excel.to_excel('sample.xlsx')     # here we converted a dataframe into excel file
```

```python
pd.read_excel('sample.xlsx',index_col=0)      # we can read excel file
```

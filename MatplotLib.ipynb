# MatplotLib
### - Visualization library (Plotting)
### - Used to create 2 D diagrams or 3 D diagrams
### - Numpy is the base library 
### - It provides object-oriented API using general purpose kits
### - Support for custom labels and texts
### - Great control of every element in figure
### - High quality o/p in many formats
### - Very customizable in general


```python
import matplotlib.pyplot as plt     # importing matplot library

#Whenever you want to display any matplotlib graphs or diagrams you need to write plt.show() but if you use percentile of matplotlib inline you don't need to write plt.show

%matplotlib inline                  
```

```python
import numpy as np
```

```python
# simple example

x=np.arange(0,10)
y=np.arange(11,21)

# plotting using matplotlibs

# plt scatter will scatter x and y values

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.scatter(x,y,c='g')    # c is color also we don't need to mention plt.show() as we already mentioned %matplotlib inline
plt.xlabel('X-axis',fontdict=font1)
plt.ylabel('Y-axis',fontdict=font2)
plt.title("2-D Graph", loc='right')

plt.grid(axis='y',color='r',linestyle = '--', linewidth = 0.5)              # we can choose which grid lines we want to display   

plt.savefig('first_plt.png')     # You can save img by right click and save img as or we can use this command
```

```python
# plot is used to draw a line through graph

plt.plot(x,y)
```

```python
y=x*x
plt.plot(x,y,'r-.')       # r-. = dashed dotted line with red color
```

```python
y=x*x*x
plt.plot(x,y,'y*',linestyle='dashed',linewidth=2, ms=10, mfc='r',mec='c')  
# there are different different markers available also mec=markeredgecolor, ms=markersize, mfc=markerfacecolor
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2-D Graph')
```

```python
# If we do not specify x-points it will take x-axis's value as 0,1,2,3,....

#y=np.arange(10,20)
y=np.array([43,65,12,45,78])      # let's take an example of array
plt.plot(y)
```

### Creating Subplots

```python
x=np.arange(0,20)
y=np.arange(21,41)
y=x*x                     

plt.subplot(2,2,1)        # we have 2 rows and 2 cols
plt.plot(x,y,'r',linestyle='dashed')
plt.subplot(2,2,2)
plt.plot(x,y,'b-.')
plt.subplot(2,2,3)
plt.plot(x,y,'y-*')
plt.subplot(2,2,4)
plt.plot(x,y,'g.')
```

## Sine and cos waves using subplot

```python
x=np.arange(0,5*np.pi,0.2)
y_sine=np.sin(x)
y_cos=np.cos(x)

plt.suptitle('WAVES')

plt.subplot(2,1,1)
plt.plot(x,y_sine,'r--') 
plt.title("Sine Wave")

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g-.')
plt.title("Cosine Wave")
```

```python
# Scatter and colormap

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')     # there are different colormaps available you can use any cmap
```

## Bar Graphs

```python
x1=[3,7,10]
y1=[4,5,7]

x2=[1,5,9]
y2=[2,6,10]

plt.bar(x1,y1,align='center')
plt.bar(x2,y2,color='k')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Bar Diagram')

''' if you want the horizontal bar then use plt.barh :
plt.barh(x1,y1)
plt.barh(x2,y2,height=0.3 )
'''
```

## Box Plot

```python
data=[np.random.normal(1,std,100) for std in range (0,4)]
plt.boxplot(data, vert=True ,patch_artist=True)

'''the circles are outliers - sometimes there is a probability that the outliers get selected while selecting a random value
, bottom lines are 0 percentile then 25, 50, 75 and 100 %  
'''
```

## Pie Chart

```python
labels='Python','C++','Java','SQL'
sizes=[345,255,285,115]
colors=['lightcoral','gold','yellowgreen','lightskyblue']
explode=(0.2,0,0,0)      # shifts the slice

plt.pie(sizes, colors=colors, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)   

# auto pct =will return float value

plt.legend(title='Languages')
```
